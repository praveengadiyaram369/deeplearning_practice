```python
import numpy as np
import tensorflow as tf

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = [[[(i+j)/100]  for i in range(5)] for j in range(100)]
y = [(i+5)/100 for i in range(100)]
```

```python
X = np.array(X, dtype = float)
y = np.array(y, dtype = float)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=11)
```

```python
rnn = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(units=1, batch_input_shape=(None, 5, 1), return_sequences=True),
    tf.keras.layers.LSTM(units=1, return_sequences=False)
])
```

```python
rnn.compile(
        optimizer='adam',
        loss='mean_absolute_error',
        metrics=['accuracy']
)
```

```python
rnn.summary()
```

```python
history = rnn.fit(X_train, y_train, epochs=400, validation_data=(X_test, y_test))
```

```python
results = rnn.predict(X_test)
```

```python
plt.scatter(range(25), results, c='r')
plt.scatter(range(25), y_test, c='b')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.show()
```
